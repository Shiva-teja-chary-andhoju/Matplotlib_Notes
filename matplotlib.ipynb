What is matplotlib? Matplotlib is a Python library used for creating static, animated, and interactive plots.
Data visualization: It's the graphical representation of data to identify patterns, trends, and insights.

```python
#pip install ipykernel in command prompt
#pip install matplotlib in command prompt
#save file as .ipynb in vscode and select the kernel in current project folder

import matplotlib.pyplot as plt

# pyplot is a collection of functions that make matplotlib work as MATLAB
```

```python
#1 Create First Chart
x = [1,2,3,4,5] # 5 values
y = [10,20,25,35,45] # 5 values

plt.plot(x,y) # line chart
plt.show() #display chart in jupyter notebook
```

```python
#2. Customize Chart

plt.plot(x,y,color = 'green',marker = '*',linestyle ='dashed',linewidth =2,markersize = 12)
plt.show()
```

```python
plt.plot(x,y,color = 'blue',marker = 'o',linestyle ='--',linewidth =1,markersize = 5)
plt.show()
```

```python
# figure size in inches
plt.figure(figsize=(4,3))

plt.plot(x,y,color = '#FF0000',marker = 'o',linestyle ='-',linewidth =3,markersize = 10) #'#FF0000 color code

plt.title("Sales per Month")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
#3. Advanced - Multiple Lines & Legends
x = [1,2,3,4,5] # 5 values
y1 = [10,20,25,35,45] # 5 values
y2 = [20,30,35,45,55]  # 5 values

#plot('xlabel','ylabel',data = obj) 
plt.plot(x,y1,label = 'Sales 2024')   # ploting y1 data  #label will be shown in the legend
plt.plot(x,y2,label = 'Sales 2025')   # ploting y2 data  #label will be shown in the legend

plt.title("YOY Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

```python
#4. Bar Chart to plot categorical data
x = ['A','B','C','D','E'] # 5 values
y = [10,20,25,35,45] # 5 values

plt.bar(x,y)
plt.show()
```

```python
#5. Histogram # to plot continuous data
# used for distribution analysis
#gaps between bins is in bar chart and no gap between bins in histogram as it is continuous data
#Data values are grouped into bins (intervals), and the height of each bar represents the frequency of observations within that interval.
# data
import random
data = [random.randint(1,20) for _ in range(500)]

plt.hist(data,bins = 10)
plt.show()
```

```python
#6. Pie chart
# used to show part-to-whole relationships/data
# data
categories = ['A','B','C','D','E'] # 5 values
sales = [10,20,25,35,45] # 5 values

plt.pie(sales,labels = categories,autopct = '%1.1f%%') #autopct to show values in percentage format
plt.title("Pie chart example")
plt.show()
```

```python
plt.pie(sales,labels = categories,autopct = '%1.1f%%',startangle = 90) #startangle by default is 0 (x-axis)
plt.title("Pie chart example")
plt.show()
```

```python
#7. Scatter plot
# used to find relationship btwn variables
# data
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]

plt.scatter(y1,y2)
plt.title("Scatter plot example")
plt.show()
```

```python
#8. Subplots
# used to show multiple charts in one figure

# data-1 - bar chart
categories = ['Mon','Tue','Wed','Thu','Fri']
sales = [10, 20, 55, 35, 45]

# data-2 - scatter plot
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]

plt.figure(figsize = (8,6))

#first plot - bar chart
plt.subplot(1,2,1) #(row,column,position)
plt.bar(categories,sales)
plt.title("Daily Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")

#second plot - scatter plot
plt.subplot(1,2,2)
plt.scatter(y1,y2)
plt.title("User Example")
plt.xlabel("User1")
plt.ylabel("User2")

plt.show()
```

```python
plt.figure(figsize = (8,6))
#first plot - bar chart
plt.subplot(2,1,1) #(row,column,position)
plt.bar(categories,sales)
plt.title("Daily Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")

#second plot - scatter plot
plt.subplot(2,1,2)
plt.scatter(y1,y2)
plt.title("User Example")
plt.xlabel("User1")
plt.ylabel("User2")

plt.show()
```

```python
#9. Matplotlib with Pandas - real data
# create df
import pandas as pd

# data
data = {
    'Month':['Jan','Feb','Mar','Apr'],
    'Sales':[12000,11000,13000,25000]
}

df = pd.DataFrame(data)
df
```

```python
plt.bar(df['Month'],df['Sales'])
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
#10. Save charts

plt.bar(df['Month'],df['Sales'])
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.savefig("Monthly_sales_fom_df.png")

plt.show()
```
